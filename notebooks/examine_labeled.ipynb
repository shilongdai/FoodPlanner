```python
import pandas as pd
from tqdm import tqdm
import config

tqdm.pandas()
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
import json

recipe_combos = []
thoughts = []
valid = []
changed = []

with open("../data/labeled_meals.json") as fp:
    records = json.load(fp)
    
for r in records:
    recipe_combos.append(r["recipes"])
    thoughts.append(r["appropriate"]["thought"])
    valid.append(r["appropriate"]["valid"])
    changed.append(r["changed"])
    
labeled_df = pd.DataFrame({
    "combo": recipe_combos,
    "thought": thoughts,
    "valid": valid,
    "changed": changed
})
labeled_df.head()
```

```python
labeled_df["predicted"] = labeled_df.apply(lambda row: row["valid"] if not row["changed"] else not row["valid"], axis=1)
labeled_df.head()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_true=labeled_df["valid"], y_pred=labeled_df["predicted"]))
```

```python
from sklearn.model_selection import train_test_split

eval, test = train_test_split(labeled_df, test_size=0.5, random_state=93, stratify=labeled_df["valid"])
test.valid.value_counts()
```

```python
from create_db import create_db_entries
import chromadb

def df_to_records(df):
    records_series = df.apply(lambda r: {
        "recipes": r["combo"],
        "appropriate": {
            "valid": r["valid"],
            "thought": r["thought"],
        },
        "changed": r["changed"]
    }, axis=1)
    return records_series.tolist()

client = chromadb.Client()
collection = client.create_collection(name=config.MEAL_VALID_EXAMPLE_COL)
db_entries = create_db_entries(df_to_records(eval), collection)
collection.add(**db_entries)
collection.count()
```

```python
from model import RandomExampleSelector
from langchain.chat_models import ChatOpenAI
from meal import ChromaMealExampleSelector
import config

config.API_KEY_PATH = "../" + config.API_KEY_PATH
config.VERBOSE = True
format_examples = "../data/format_examples.json"
with open(config.API_KEY_PATH, "r") as fp:
    API_KEY = fp.read().strip()
plan_llm = ChatOpenAI(openai_api_key=API_KEY, temperature=config.MEAL_VALID_TEMP,
                          model_name=config.MEAL_VALID_MODEL, model_kwargs=dict(timeout=config.TIMEOUT))
with open(format_examples, "r") as fp:
    examples = json.load(fp)
    selector = RandomExampleSelector(examples, k=2)

meal_selector = ChromaMealExampleSelector(collection=collection)
meal_selector.select_examples({"recipes": "lobster"})
```

```python
from meal import RecipeMealInput
from meal import evaluate_recipes

def predict_valid(recipe_record):
    recipes = []
    for r in recipe_record:
        recipes.append(RecipeMealInput(**r))
    model_output = evaluate_recipes(plan_llm, recipes, selector, meal_selector)
    return model_output["thought"], model_output["valid"]

predict_valid(labeled_df["combo"].iloc[0])
```

```python
print(classification_report(y_true=test["valid"], y_pred=test["predicted"]))
```

```python
config.VERBOSE = False
outputs = test.combo.progress_apply(predict_valid).tolist()
new_thought, new_valid = zip(*outputs)
test["new_thought"] = new_thought
test["new_predicted"] = new_valid
test.head()
```

```python
print(classification_report(y_true=test["valid"], y_pred=test["new_predicted"]))
```
